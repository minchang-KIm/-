```python
# 필요한 모듈 로딩

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time


# 크롬브라우저 실행

#Step 3. 크롬 드라이버 설정 및 웹 페이지 열기
#크롬드라이버 경로 설정
s = Service("c:/py_temp/chromedriver.exe")

# 크롬드라이버 실행
driver = webdriver.Chrome(service=s)

# Riss 사이트 접속
driver.get("http://www.riss.or.kr/index.do#search")

driver.maximize_window() # 크롬 창 최대화(전체화면)

time.sleep(2) # 컴퓨터가 ()에 있는 시간만큼(초) 기다린다.
#검색창 지정
search_bar = driver.find_element(By.ID, 'query')
search_bar.send_keys("낮잠")

# 검색버튼 클릭
driver.find_element(By.XPATH, '//*[@id="in"]/div/button').click()

driver.find_element(By.XPATH, '//*[@id="tabMenu"]/ul/li/div/ul/li[2]/a').click()

time.sleep(2)
```

```python
from bs4 import BeautifulSoup
html_1 = driver.page_source
# 셀레늄한테 페이지 소스 들고와서 html_1 이라는 통에 담으라고 시킴


soup = BeautifulSoup(html_1, 'html.parser')
```

```python
content = soup.find('div', 'srchResultListW').find('li')
# find_all 은 list

no =0
#indent hotkey : ctrl + []
for i in content:
    try: # 일단 아래코드 돌려보고
        제목 = i.find('p', 'title')
        print("제목 :", 제목.get_text())
        no += 1
        abstract = i.find('p', 'preAbstract')
        저자 = i.find('span', 'writer')
        print(no)
        print("저자 :", 저자.get_text())
        print("내용 :", abstract.get_text())
        print("="*100)
    except : #에러나면 얘를 실행해라  <= 우리눈에 안 보이는 비어있는 li 때문.
        pass
```

```python
content = soup.find('div', 'srchResultListW').find_all('li')
# find_all 은 list형태로 결과물을 가져온다.

# 리스트 선언
no_list = []
title_list = []
body_list = []
author_list = []

no =0
#indent hotkey : ctrl + []
for i in content:
    try: # 일단 아래코드 돌려보고

        제목 = i.find('p', 'title').get_text()
        try:
            본문 = i.find('p', 'preAbstract').get_text()
        except:
            본문 = "본문없음"
        저자 = i.find('span', 'writer').get_text()

        no += 1
        print(no)
        no_list.append(no)
        print("제목 :", 제목)
        title_list.append(제목)

        print("내용 :", 본문)
        body_list.append(본문)

        print("저자 :", 저자)
        author_list.append(저자)

        print("="*100)


    except : #에러나면 얘를 실행해라  <= 우리눈에 안 보이는 비어있는 li 때문.
        pass
```

```python
import pandas as pd # pandas 라는 모듈을 pd라는 이름으로 불러와라

df = pd.DataFrame() # df라는 이름의 DataFrame(테이블)을 만들어라
# 데이터프레임이름['변수이름'] = 리스트이름
df['no'] = no_list
df['제목'] = title_list
df['저자'] = author_list
df['본문'] = body_list
df
```
